```julia
calc_dir = "/home/ibs/GRASP2018/work/Ni"
state="Ni_even";
```

## RNUCLEUS calc

```julia
rnucleus= RnucleusCmd.Rnucleus(calc_dir=calc_dir, state=state, Z=28, A=61, neutralMass=58.6934, I=1, NDM=1, NQM=1)
```

```julia
Basics.Execute(rnucleus)
```

## Rcsfgenerate calc

```julia
configurations= [
                "2s(2,i)2p(6,i)3s(1,i)3p(4,i)",           # even
                "2s(2,i)2p(6,i)3s(2,i)3p(2,i)3d(1,i)",    # even
                # "2s(2,i)2p(6,i)3s(2,i)3p(3,i)"            # odd
                ]
```

```julia
configs = RcsfgenerateCmd.Configurations(core="He", Configs=configurations)
```

```julia
active_orbitals="3s,3p,3d"
rcsfgenerate = RcsfgenerateCmd.Rcsfgenerate(
    calc_dir=calc_dir,
    Configs=configs,
    Jl=1,
    Jh=9,
    noex=0,
    active_orbitals=active_orbitals
)
```

```julia
Basics.Execute(rcsfgenerate)
```
