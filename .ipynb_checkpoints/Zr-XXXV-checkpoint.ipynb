# Zr-XXXV using JuliaGrasp Package (Fatma paper)

```julia
import Pkg
pakg_Path =joinpath(@__DIR__);
Pkg.activate(pakg_Path);
# Pkg.add(url="https://github.com/ishaarany/JuliaGrasp.jl")
using JuliaGrasp
```

## Initialization

```julia
principle_orbital= 3
n_max=principle_orbital+1;
active_orbitals="4s,4d";
excitations=0; 
radial_wavefunctions = excitations > 0 ? 1 : 2;
Jl=0;
Jh=10;

default= DefaultModule.Default(
        calc_dir= "/home/ibs/GRASP2018/work/Zr-XXXV",
        state= "even",
        principle_orbital= principle_orbital,
        state_folder= "",
        out_folder= excitations > 0 ? "sd" : "mr",
        in_folder= excitations >  0 ? "sd" : "mr",
        radial_wavefunctions = radial_wavefunctions,
        n_max = n_max
    )
default= DefaultModule.Init(default)
```

```julia
rnucleus= RnucleusCmd.Rnucleus(default=default, Z=40, A=92, neutralMass=91.224, I=0, NDM=0, NQM=0);
```

```julia
# even:  2s2 2p2  , 2p4 ,  2s2 2p 3p , 2s 2p2 3s, 2s 2p2 3d, 2p3 3p
# odd:  2s 2p3 , 2s2 2p 3s, 2s2 2p 3d, 2s 2p2 3p, 2p3 3s,2p3 3d   

configurations= [
                # even
                "2s(2,*)2p(2,*)",              
                "2p(4,*)",
                "2s(2,*)2p(1,*)3p(1,*)",
                "2s(1,*)2p(1,*)3p(1,*)4s(1,*)",
                # "2s(1,*)2p(1,*)3p(1,*)4d(1,*)",
                "2s(1,*)2p(2,*)3s(1,*))",
                "2s(1,*)2p(2,*)3d(1,*)",
                "2p(3,*)3p(1,*))",
                # "2s(2,*)3d(2,*)",
    
                #odd
                # "2s(1,*) 2p(3,*)" ,                        
                # "2s(2,*)2p(1,*)3s(1,*)",
                # "2s(2,*)2p(1,*)3d(1,*)",
                # "2s(1,*)2p(2,*)3p(1,*)", 
                # "2p(3,*)3s(1,*)", 
                # "2p(3,*)3d(1,*)", 
                ];
```

```julia
configs = RcsfgenerateCmd.Configurations(core="He", Configs=configurations);
```

```julia
rcsfgenerate = RcsfgenerateCmd.Rcsfgenerate(
    default=default,
    Configs=configs,
    Jl=Jl,
    Jh=Jh,
    noex=excitations,
    active_orbitals=active_orbitals
);
```

```julia
calc= CalcCmd.Calc(
    default       =default,
    rnucleus      =rnucleus,
    rcsfgenerate  =rcsfgenerate,
    # rangular      =rangular,
    # rcsfinteract  =rcsfinteract,
    # rwfnestimate  =rwfnestimate,
    # rmcdhf        =rmcdhf,
    # rci           =rci,
    # jj2lsj        =jj2lsj,
    # rlevels       =rlevels
    );
```

```julia
CalcCmd.Execute(calc)
```
