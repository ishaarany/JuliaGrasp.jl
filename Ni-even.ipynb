# Sixth Grasp Example Ni-even using JuliaGrasp Package

```julia
import Pkg
pakg_Path =joinpath(@__DIR__);
Pkg.activate(pakg_Path);
using JuliaGrasp
```

## Initialization

```julia
principle_orbital= 3
n_max=4;
active_orbitals="3s,3p,3d";
excitations=2; 
radial_wavefunctions = excitations > 0 ? 1 : 2;
Jl=1;
Jh=9;

default= DefaultModule.Default(
        calc_dir= "/home/ibs/GRASP2018/work/Ni",
        state= "even",
        principle_orbital= principle_orbital,
        state_folder= "",
        out_folder= excitations > 0 ? "sd" : "mr",
        in_folder= excitations >  0 ? "sd" : "mr",
        radial_wavefunctions = radial_wavefunctions,
        n_max = n_max
    )
default= DefaultModule.Init(default)
```

```julia
rnucleus= RnucleusCmd.Rnucleus(default=default, Z=28, A=61, neutralMass=58.6934, I=1, NDM=1, NQM=1);
```

```julia
configurations= [
                "2s(2,i)2p(6,i)3s(1,i)3p(4,i)",            # even
                "2s(2,i)2p(6,i)3s(2,i)3p(2,i)3d(1,i)",     # even
                # "2s(2,i)2p(6,i)3s(2,i)3p(3,i)"            # odd
                ];
```

```julia
configs = RcsfgenerateCmd.Configurations(core="He", Configs=configurations);
```

```julia
rcsfgenerate = RcsfgenerateCmd.Rcsfgenerate(
    default=default,
    Configs=configs,
    Jl=Jl,
    Jh=Jh,
    noex=excitations,
    active_orbitals=active_orbitals
);
```
